## Customer Segmentation Analysis


```python
import pandas as pd
```

```python
df_orders = pd.read_excel('orders.xlsx', sheet_name='Sheet1')
```

```python
df_bikeshops = pd.read_excel('bikeshops.xlsx', sheet_name = 'Sheet1')
```

```python
df_bikes = pd.read_excel('bikes.xlsx', sheet_name = 'Sheet1')
```

```python
df_orders.head()
```

```python
df_bikeshops.head()
```

```python
df_bikes.head()
```

```python
df_merged = df_orders.merge(df_bikeshops, left_on = 'customer.id', right_on= 'bikeshop.id')
```

```python
df_merged.head()
```

```python
df_merged = df_merged.merge(df_bikes, left_on= 'product.id', right_on = 'bike.id')
```

```python
df_merged.head()
```

Removing duplicate columns

```python
df_merged = df_merged.drop(['customer.id'], axis=1)
```

```python
df_merged = df_merged.drop(['product.id'], axis=1)
```

```python
df_merged.head()
```

```python
df_merged['price'] = df_merged['price'].apply(lambda x: 'high' if x > df_merged['price'].median() else 'low')
```

```python
df_merged.head()
```

```python
df_grouped = df_merged.groupby(by=['bikeshop.name', 'model', 'category1', 'category2', 'frame', 'price'], as_index=False).agg({'quantity': 'sum'})
```

```python
bikeshop_names = df_grouped['bikeshop.name'].unique()
```

```python
df_grouped.head()
```

```python
model_pivot = pd.pivot_table(df_grouped, index=['model', 'category1', 'category2', 'frame', 'price'], columns = ['bikeshop.name'], values='quantity', fill_value=0)
```

```python
model_pivot.head()
```

Normalizing individual columns

```python
for i in model_pivot.columns:
    model_pivot.loc[:,i] = model_pivot.loc[:,i].apply(lambda x: x/model_pivot[i].sum())
```

```python
model_pivot.T.head()
```

So we are clustering different bike models based on their quantity purchased (across different bike shops).

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import numpy as np
```

In order to obtain a optimal value for the number of clusters, I use silhouette analysis. Silhouette analysis can be used to study the separation distance between the resulting clusters. 

```python
cluster_sizes = [i for i in range(4,9)]
```

### k-means

```python
best_k = None
best_score = -np.inf
for k in cluster_sizes:
    kmeans = KMeans(n_clusters=k, random_state=90)
    labels = kmeans.fit_predict(model_pivot.T)
    silhouette_vals = silhouette_samples(model_pivot.T, labels)
    avg_score = np.mean(silhouette_vals)
    
    print('Cluster size: {}, Score: {}'.format(str(k), str(avg_score)))
    if avg_score > best_score:
        best_k = k
        best_score = avg_score
print('Optimal Cluster size: ', best_k)
```

```python
result = KMeans(n_clusters=best_k, random_state=90)
```

```python
labels = result.fit_predict(model_pivot.T)
```

Printing customer names in each segment

```python
print('Bikeshop Name', ':', 'Cluster Number')
clusters = {}
for idx, i in enumerate(labels):
    if i in clusters:
        clusters[i].append(model_pivot.T.index[idx])
    else:
        clusters[i] = []
        clusters[i].append(model_pivot.T.index[idx])
```

### Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering

clustering = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
cluster_assignments = clustering.fit_predict(model_pivot.T)
```

```python

clusters_new = {}
for idx, i in enumerate(cluster_assignments):
    if i in clusters_new:
        clusters_new[i].append(model_pivot.T.index[idx])
    else:
        clusters_new[i] = []
        clusters_new[i].append(model_pivot.T.index[idx])
```
